```python
import pandas as pd
```

```python
data = [[6, 'Alice'], [2, 'Bob'], [7, 'Alex']]
users = pd.DataFrame(data, columns=['user_id', 'user_name']).astype({'user_id':'Int64', 'user_name':'object'})
data = [[215, 6], [209, 2], [208, 2], [210, 6], [208, 6], [209, 7], [209, 6], [215, 7], [208, 7], [210, 2], [207, 2], [210, 7]]
register = pd.DataFrame(data, columns=['contest_id', 'user_id']).astype({'contest_id':'Int64', 'user_id':'Int64'})
```

```python
register
```

```python
users
```

```python
ryt_join = register.merge(users, on="user_id", how="right")
```

```python
ryt_join
```

```python
total_users = ryt_join["user_id"].nunique()
users_per_contest = ryt_join.groupby("contest_id")["user_id"].count().reset_index()
users_per_contest
```

```python
users_per_contest["percentage"] = (users_per_contest["user_id"] * 100 / total_users).round(2).dropna()
```

```python
users_per_contest = users_per_contest.sort_values(["percentage", "user_id"], ascending=[False, True])
```
